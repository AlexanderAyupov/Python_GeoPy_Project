```python
def distancer(row):
    coords_1 = (row['USER_LAT'], row['USER_LONG'])
    coords_2 = (row['VENUE_LAT'], row['VENUE_LONG'])
    return geopy.distance.geodesic(coords_1, coords_2).km
df['USER_VENUE_DISTANCE'] = df.apply(distancer, axis=1)
```

# Cross Validation
### Comparing the performance of the models


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
```

### Importing models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
X=df[['ITEM_COUNT', 'USER_LAT', 'USER_LONG', 'VENUE_LAT', 'VENUE_LONG',
       'ESTIMATED_DELIVERY_MINUTES','CLOUD_COVERAGE', 'TEMPERATURE', 'WIND_SPEED', 'PRECIPITATION',
       'USER_VENUE_DISTANCE','MONTH', 'DAY', 'HOUR', 'CLUSTERS',
       'CLUSTERS_CONSTRAINED','clusters_on_target']]
y=df['TARGET']
```

```python
df['TARGET'].value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
# list of models
models = [LogisticRegression(max_iter=1000), SVC(kernel='linear'), KNeighborsClassifier(), RandomForestClassifier()]
```

```python
def compare_models_train_test():

  for model in models:

    # training the model
    model.fit(X_train, y_train)

    # evaluating the model
    test_data_prediction = model.predict(X_test)

    accuracy = accuracy_score(y_test, test_data_prediction)

    print('Accuracy score of the ', model, ' = ', accuracy)
```

```python
compare_models_train_test()
```

### Creating a Function to compare the models

```python
# list of models
models = [LogisticRegression(max_iter=1000), SVC(kernel='linear'), KNeighborsClassifier(), RandomForestClassifier()]
```

```python
def compare_models_cross_validation():

  for model in models:

    cv_score = cross_val_score(model, X,Y, cv=5)

    mean_accuracy = sum(cv_score)/len(cv_score)

    mean_accuracy = mean_accuracy*100

    mean_accuracy = round(mean_accuracy, 2)

    print('Cross Validation accuracies for ', model, '=  ', cv_score)
    print('Accuracy % of the ', model, mean_accuracy)
    print('----------------------------------------------')
```

```python
compare_models_cross_validation()
```
